```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#DataViz
import seaborn as sns
```

## Dataset exploration

```python
df = pd.read_csv('../data/GHISACONUS_2008_001_speclib.csv')
```

```python
df.describe()
```

In addition to 242 bands (`X427` to `X2395`), each of 10 nm bandwidth in the 400-2500 nm range, the dataset includes:

- `Unique ID` : identifyig a unique crop field
- `XAEZ` : identifying the Agrogeological Zone (all in USA)
- `Month` : month in which the image was acquired
- `Year` : year in which the image was acquired
- `jd` : Julian day in which the image was acquired (used to identify crop stage)
- `long` : longitude
- `lat` : latitude
- `Crop` : type of crop, cross referenced with *USDA Cropland Data Layer*.
- `Stage` : crop's stage of growth, cross referenced with *Center for Sustainability and the Global Environment*

```python
df.isnull().sum()
```

Reflectance has been consistently removed in four portions of the spectrum have been removed. I transform the nulls to `0` in this notebook to visualize the available spectrum in its integrity. In modeling datasets I will drop these columns.

```python
df = df.fillna(value = 0)
```

Unique ID is different for each row, I will set that as an index.

```python
df.set_index('UniqueID', inplace=True)
```

Because all data come from the `USA`, I drop this column in the analysis.

```python
df.drop(columns = ['Country'], inplace= True)
```

## Crop Field Features Analysis 

```python
def custom_pallete(custom_colors):
    customPalette = sns.set_palette(sns.color_palette(custom_colors))
    sns.palplot(sns.color_palette(custom_colors),size=0.8)
    plt.tick_params(axis='both', labelsize=0, length = 0)
```

```python
contrast =["#E26D5C","#FFCD70", "#38618C", '#723D46', '#472D30']
vibrant =["#E3170A","#A9E5BB", "#FCF6B1", '#F7B32B', '#2D1E2F']
custom_pallete(contrast)
```

```python
plt.figure(figsize=(10,10))
sns.stripplot(x=df['Month'], y=df['Crop'], alpha=.05)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);
#, color = ['Stage']
```

From the graph above we see that the crop classes are not equally distributed across the 5 months represented in the dataframe. We can also see that `Rice` in particular seems strongly underrepresented. I decide then to check value counts for Crop and Stage.

```python
df.groupby('Stage')['Crop'].value_counts()
```

I notice a strong class imbalance.

Also, I want to see how the data are distributed in the space. I use latitude and longitude as elements in a scatterplot to have a quick view.

```python
plt.figure(figsize=(30,15))
sns.scatterplot(x=df['long'], y=df['lat'], hue = df["Crop"], alpha=0.1, markers=df["Crop"]);
```

Regions of data collections appear to be 6 and well distinct. Although, each crop type is represented in more than one region, and no region is referred a unique data type. 

## Sprectral Features Analysis

Moving into specrtum feature analysis, I plot two random elements to get a vew of the level of detail represented.

As mentioned I notice the areas turned to 0 and the first portion of the spectrum seems to be the biggest overall.

### Crop
Creating dataframes with subsects of different crops

```python
corn_df = df[df['Crop'] == 'corn']
soy_df = df[df['Crop'] == 'soybean']
rice_df = df[df['Crop'] == 'rice']
cotton_df = df[df['Crop'] == 'cotton']
weat_df = df[df['Crop'] == 'winter_wheat']
```

```python
plt.figure(figsize=(10,5))

soy_df.iloc[0].filter(like='X', axis=0).plot(kind='line') 
corn_df.iloc[0].filter(like='X', axis=0).plot(kind='line') 
rice_df.iloc[0].filter(like='X', axis=0).plot(kind='line')
cotton_df.iloc[0].filter(like='X', axis=0).plot(kind='line')
weat_df.iloc[0].filter(like='X', axis=0).plot(kind='line')

#Legend
plt.legend(['soy', 'corn', 'rice', 'cotton', 'weat'])
#Title
plt.title('Crop Spectra Samples');
```

There are all the crops (in different states of their development). we can see a distinctive difference in the bands between 427 and 930.

Will consider this in running an experiment.

### Stage

I decide to preliminarly plot the spectra of soy in different stages, to check for differences.

```python
plt.figure(figsize=(10,5))
soy_df.iloc[1].filter(like='X', axis=0).plot(kind='line') # Critical
soy_df.iloc[8].filter(like='X', axis=0).plot(kind='line') # Mature
soy_df.iloc[19].filter(like='X', axis=0).plot(kind='line') # Harvest
soy_df.iloc[49].filter(like='X', axis=0).plot(kind='line'); #Emerge_VEarly

#Legend
plt.legend(['critical', 'mature', 'harvest', 'early'])
#Title
plt.title('Soy Spectra Samples Per Stage');
```

## Conclusions

From this preliminary analysis and data overview we have an understanding of the features we could leverage on for modeling. In particular, other than the spectra elements, I decide to:
- not use `month` in the modeling, because it might be redundent with stage  as we are deriving stage from a library that uses Julian days. Also some crops have been captured in one month only, creating a strong correlation and potentially inserting bias in the modeling.

- potentially I could use the `ageroecological zone` as a categorical feature within the models, but considering my objective is to create a model that can be applied outside of the USA and with a broad global application, I would try not to use it.
